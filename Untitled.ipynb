```python
df = pd.read_csv('ajay.csv', header=None)
df.columns = ['variance', 'skewness', 'kurtosis', 'entropy', 'authenticate']
df.head()
```

```python
df.shape
```

```python
sns.pairplot(df, hue='authenticate')
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x=df['authenticate'])
target_count = df.authenticate.value_counts()
plt.annotate(s=target_count[0], xy=(-0.04,10+target_count[0]), size=10)
plt.annotate(s=target_count[1], xy=(0.96,10+target_count[1]), size=10)
plt.show()
```

```python
balanced = target_count[0] - target_count[1]
df = df.sample(frac=1, random_state=42).sort_values(by='authenticate')
df = df[balanced:]
print(df['authenticate'].value_counts())
```

```python
x = df.drop(["authenticate"],axis=1)
y = df["authenticate"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
scalar = StandardScaler()
scalar.fit(x_train)
x_train = scalar.transform(x_train)
x_test = scalar.transform(x_test)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
ypred = model.predict(x_test)
accuracy_score(y_test, ypred)
```
